#### `multiplying 1000 where KM_Representation == k and multiplying 100000 where KM_Representation is L`

```python
def convert_km(row):
    if row['KM_Representation'] == 'k':
        return int(row['KM_Driven'] * 1000)
    elif row['KM_Representation'] == 'L':
        return int(row['KM_Driven'] * 100000)
    else:
        return row['KM_Driven']  # Leave unchanged if unknown representation

data['KM_Driven_Actual'] = data.apply(convert_km, axis=1)
```

```python
data['Price'] = data['Price'].str.extract("(\d.\d+)").astype(float)
```

```python
def price(row):
    
    return int(row['Price'] * 100000)
    

data['Price'] = data.apply(price, axis=1)
```
